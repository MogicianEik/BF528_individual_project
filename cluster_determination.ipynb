```python
import pandas as pd
import seaborn as sns
from scipy import stats
import scipy
from scipy.stats import rankdata
```

```python
df = pd.read_csv('CEL_files/write.csv', header=0, index_col=0)
df.head()
```

```python
meta = pd.read_csv('proj_metadata.csv', header=0)
dic = {}
for index, row in meta.iterrows():
    file = row['CEL_path'].split('/')[-1][:9]
    dic[file] = row['cit-coloncancermolecularsubtype']
```

```python
ax = sns.clustermap(df, row_cluster=False, cmap="mako",method='weighted',metric = 'correlation',figsize = (40,50))
for tick_label in ax.ax_heatmap.axes.get_xticklabels():
    tick_text = tick_label.get_text()
    if dic[tick_text[:9]] == 'C3':
        tick_label.set_color('red')
    else:
        tick_label.set_color('blue')
```

```python
ax.savefig('clustermap.png')
```

```python
den = scipy.cluster.hierarchy.dendrogram(ax.dendrogram_col.linkage,
                                         labels = df.columns,
                                         color_threshold=0.12)  
from collections import defaultdict

def get_cluster_classes(den, label='ivl'):
    cluster_idxs = defaultdict(list)
    for c, pi in zip(den['color_list'], den['icoord']):
        for leg in pi[1:3]:
            i = (leg - 5.0) / 10.0
            if abs(i - int(i)) < 1e-5:
                cluster_idxs[c].append(int(i))

    cluster_classes = {}
    for c, l in cluster_idxs.items():
        i_l = [den[label][i] for i in l]
        cluster_classes[c] = i_l

    return cluster_classes

clusters = get_cluster_classes(den)

clusters
```

```python
print(len(clusters['C1']))
print(len(clusters['C2']))
```

```python
df1 = df[clusters['C1']]
df2 = df[clusters['C2']]
```

```python
new_df = {}
for index,row in df1.iterrows():
    results = stats.ttest_ind(row, df2.loc[index], equal_var = False)
    new_df[index] = [results[0],results[1]]
```

```python
de = pd.DataFrame.from_dict(new_df, orient='index',
                       columns=['t-statistic', 'p-value'])
```

```python
def fdr(p_vals):
    ranked_p_values = rankdata(p_vals)
    fdr = p_vals * len(p_vals) / ranked_p_values
    fdr[fdr > 1] = 1

    return fdr
fdrs = fdr(de['p-value'])
de['fdr'] = fdrs
```

```python
de.loc[de['fdr'] < 0.05].sort_values(by=['fdr'],ascending=True).head(20)
```
